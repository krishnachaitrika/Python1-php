```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
x = [1,2,3,4]
y = [5,6,7,8]
plt.plot(x,y)
plt.grid()
plt.show()
```

```python
#pyplot API
#univarient - numerical
data = {
 "salary" : [45000,52000,61000,48000,75000,39000,82000,56000,47000,68000,91000,51000,60000,72000,46000,84000,55000,95000,43000,77000]
}
```

```python
df = pd.DataFrame(data)
print(df)
```

```python
#line plot
plt.plot(df["salary"],color= "red", marker = '*', linestyle = ":",linewidth = 2)
plt.grid()
plt.show()
```

```python
#histogram 
plt.hist(df["salary"],color= 'green',bins = 5)
plt.grid()
plt.show()
```

```python
#box plot
plt.boxplot(df["salary"])
plt.show()
```

```python
#univarient = string
data2 = {
    "Salary": [
        45000, 52000, 61000, 48000, 75000,
        39000, 82000, 56000, 47000, 68000,
        91000, 51000, 60000, 72000, 46000,
        84000, 55000, 95000, 43000, 77000
    ],
    "Department": [
        "HR", "IT", "Finance", "IT", "HR",
        "Sales", "Finance", "IT", "Sales", "Finance",
        "IT", "HR", "Sales", "Finance", "IT",
        "Finance", "HR", "Sales", "IT", "Finance"
    ]
}
```

```python
df2 = pd.DataFrame(data2)
print(df2)
```

```python
#pie char
count = df2['Department'].value_counts()
count
```

```python
plt.pie(count, labels = count.index,autopct = "%1.1f")
plt.show()
```

```python
#countplot
plt.bar(count.index,count,color = ['red','green','blue','yellow'])
plt.show()
```

```python
#bivarient 
df2["age"]= [
        25, 30, 35, 28, 40,
        26, 38, 29, 32, 41,
        36, 27, 34, 39, 31,
        37, 33, 42, 24, 45
    ]
```

```python
df2.head()
```

```python
#scatter plot
plt.scatter(df2["age"],df2["Salary"],color = 'orange')
plt.show()
```

```python
sort_age = df2.sort_values("age")
```

```python
#line plot
plt.plot(sort_age["age"],df2["Salary"] ,marker ="o")
plt.show()
```

```python
#bar chat
plt.bar(sort_age["age"],df2["Salary"],color= 'pink')
plt.show()
```

```python
plt.plot(sort_age["age"],df2["Department"],color = 'yellow',marker = 'o')
plt.show()
```

```python
hr_salary = df2[df2["Department"] == "HR"]["Salary"]
hr_salary
```

```python
fin_salary = df2[df2["Department"] == "Finance"]["Salary"]
fin_salary
```

```python
IT_salary = df2[df2["Department"] == "IT"]["Salary"]
IT_salary
```

```python
sales_salary = df2[df2["Department"] == "Sales"]["Salary"]
sales_salary
```

```python
plt.boxplot([hr_salary, IT_salary, sales_salary, fin_salary])
plt.xticks([1, 2, 3, 4], ["HR", "IT", "Sales", "Finance"])  
plt.show()
```

```python
#group by is used club the total sum of salary
salary_dept = df2.groupby("Department")["Salary"].sum()
salary_dept
```

```python
plt.pie(salary_dept,labels= salary_dept.index,autopct="%1.1f")
plt.show()
```

```python
#multivarient
df2["exp"]=[1,2,3,4,5,5,3,2,3,1,4,5,2,7,8,5,3,1,3,2 ]
df2.head()
```

```python
#bubble plot
plt.scatter(df2["age"],df2["Salary"],s = df2["exp"]*50,color = "pink",edgecolors = "black")
plt.show()
```
