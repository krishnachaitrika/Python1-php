```python
#sum of digits
def sum_of_digit(n):
    sum = 0
    while n!=0:
        digit = n%10
        sum = digit + sum
        n//=10
    return sum
b = int(input())
print(sum_of_digit(b))
```

```python
#reverse a number
def reverse_no(n):
    rev = 0
    while n!=0:
        digit = n%10
        rev = digit+rev*10
        n//=10
    return rev

x = int(input())
print(reverse_no(x))        
```

```python
#reverse string
def reverse_string(a):
    rev = " "
    rev = a[::-1]
    return rev

b = input()
print(reverse_string(b))
```

```python
#factorial
def factorial(n):
    fact = 1
    for i in range(1,n+1):
        fact = fact*i
    return fact
b = int(input())
print(factorial(b))
```

```python
#panlindrome(Number)
def palindrome(n):
    original = n
    rev = 0
    while n!=0:
        digit = n%10
        rev = digit + rev*10
        n//=10
    if (original == rev):
        return "Number is a palindrome"
    else:
        return "Number is not a palindrome"


b = int(input())
print(palindrome(b))
```

```python
#string palindrome
def palindrome(n):
    left = 0
    right = len(n)-1
    while left<right:
        if n[left]!=n[right]:
            return "its not a palindrome"
        left+=1
        right-=1
    return "its a palindrome"

b=input()
print(palindrome(b))
```

```python
#max number
def max_number(n):
    largest = 0 
    x = len(n)
    for i in range(x):
        if n[i] > largest:
            largest = n[i]
    return largest
b = list(map(int,input().split()))
print(max_number(b))
```

```python
#secound largest
def max2_number(n):
    largest = 0 
    second_largest = 0
    for num in n:
        if num > largest:
            second_largest = largest
            largest = num
        elif num > second_largest and num != largest:
            second_largest = num

    return second_largest
b = list(map(int,input().split()))
print(max2_number(b))
```

```python
#fibbnocie series
def feb_series(n):
    a = 0
    b = 1
    for i in range(1,n+1):
        print(a, end=" ")
        a, b = b, a + b
    return a
b = int(input())
print(feb_series(b))
```

```python
#Amstrong number
def amstrong_no(n):
    original = n
    power = len(str(n))
    am = 0
    while n!=0:
        digit = n%10
        am += digit**power
        n//=10
    if am == original:
        return "Armstrong"
    else:
        return "Not Armstrong"

b = int(input())   
print(amstrong_no(b))
```

```python
#reverse array
def reverse_array(arr):
    new = []
    for i in range(len(arr)-1,-1,-1):
        new.append(arr[i])
    return new
b = list(map(int,input().split()))
print(reverse_array(b))
```

```python
#pointers reverse array
def reverse_arr(arr):
    left = 0
    right = len(arr)-1
    while left<right:
        arr[right],arr[left]=arr[left],arr[right]
        left +=1
        right-=1
    return arr
b = list(map(int,input().split()))
print(reverse_arr(b))
```

```python
#frequency count
def frequency_count(arr):
    s = {}
    for i in arr:
        if i in s:
            s[i]+=1
        else:
            s[i]=1
    return s
b = list(map(int,input().split()))
print(frequency_count(b))
```

```python
#remove white spaces
def remove_whitespace(s):
    result = ""
    for ch in s:
        if ch != " ":
            result += ch
    return result
n = input()
print(remove_whitespace(n))
```

```python
#print vowels
def vowels(s):
    u = []
    for i in s:
        if (i == 'a'or i == 'e' or i == 'i' or i == 'o'or i == 'u'):
            u.append(i)
    return u
n = input()
print(vowels(n))
```

```python
#count vowels
def vowels(s):
    count = 0
    for i in s:
        if i.lower() in "aeiou":
           count+=1
    return count
n = input()
print(vowels(n))
```

```python
#sum of natural numbers
def sum_natural(n):
    total = 0
    if n==0 or n==1:
        return 0
    else:
        for i in range(1,n+1):
            total += i
        return total
b = int(input())
print(sum_natural(b))
```

```python
#consonants
def consonant(n):
    count = 0
    for ch in n:
        if ch.lower()!="aeiou":
            count+=1
    return count
a = input()
print(consonant(a))
```

```python
#GCD and LCM
def gcd(a,b):
    while b!=0:
        a,b = b,a%b
    return a
n = int(input())
m = int(input())
print(gcd(n,m))
```

```python
def lcm(a,b):
    return a*b//gcd(a,b)
n = int(input())
m = int(input())
print("lcm:", lcm(n,m))
```
