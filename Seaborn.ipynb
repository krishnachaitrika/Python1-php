```python
#palette
sns.set_context("notebook")
sns.set_style("darkgrid")
sns.scatterplot(data=peg, x="flipper_length_mm", y="body_mass_g",hue = 'island',palette='Set2')
sns.despine()
```

```python
#two different 
sns.scatterplot(data=peg, x="flipper_length_mm", y="body_mass_g",hue = 'island',palette='Set2',style = "sex")
sns.despine()
```

```python
#strip-plot
sns.stripplot(data=peg, x="species", y="body_mass_g",hue = 'island',jitter = True)
```

```python
#swarn-plot
sns.set_context("talk")
sns.swarmplot(data=peg, x="species", y="body_mass_g",hue = 'island',palette='Set3')
```

```python
sns.set_context("paper")
sns.histplot(data=peg, x="body_mass_g",hue = "sex")
```

```python
#regression plot
sns.set_style("whitegrid")
sns.regplot(data=peg,x="body_mass_g",y="flipper_length_mm",color='pink')
```

```python
#lineplot
sns.lineplot(data=peg,x="body_mass_g",y="flipper_length_mm",color='blue',hue = 'sex',style='island')
```

```python
#joint-plots
sns.jointplot(data=peg,x="body_mass_g",y="flipper_length_mm",hue ="sex")
```

```python
#bar-plot
sns.barplot(data=peg,x="species",y="body_mass_g",hue = 'sex',color='pink',estimator = np.median)
```

```python
#box-plot
sns.boxplot(data = peg,x='body_mass_g',y='species',hue = 'island')
```

```python
#voline plot
sns.violinplot(data=peg,x='species',y = 'body_mass_g',hue = 'sex')
```

```python
#kde plot
sns.kdeplot(data=peg,x='body_mass_g',hue='species',palette='Set2',fill = True)
```

```python
#heatmap
columns = ['bill_length_mm','bill_depth_mm','body_mass_g']
sns.heatmap(data=peg[columns].corr(),annot = True , cmap = 'Blues')
```

```python
#rug-plot
sns.rugplot(data=peg ,x='body_mass_g',hue='species',palette='Set2',height=0.2)
```

```python
#pairplot
sns.pairplot(data=peg,hue = 'sex',palette='Set2')
```

```python
#pair-grid
g = sns.pairplot(data=peg,hue = 'sex',palette='Set2')
g.map_upper(sns.scatterplot)
g.map_lower(sns.kdeplot)
g.map_diag(sns.histplot)
```
